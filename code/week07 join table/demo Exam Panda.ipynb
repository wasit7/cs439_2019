```python
import sqlite3
import pandas as pd
con = sqlite3.connect('db.sqlite3')
df = pd.read_sql_query("SELECT * FROM sqlite_master", con)
df
```

```python
df_member = pd.read_sql_query("SELECT * FROM myapp_member", con)
df_room = pd.read_sql_query("SELECT * FROM myapp_room", con)
df_rent = pd.read_sql_query("SELECT * FROM myapp_rent", con, parse_dates=['start','stop'])
```

```python
df_member.to_csv('member.csv')
df_room.to_csv('room.csv')
df_rent.to_csv('rent.csv')
```

```python
df_member.head()
```

```python
df_rent.head()
```

```python
df=df_member.merge(df_rent,how='inner',left_on='id',right_on='member_id')\
    .merge(df_room,left_on='room_id',right_on='id')
```

```python
df.shape
```

```python
# Q1 Who did spend most money for renting?
df.groupby('name').agg({'cost':'sum'})\
    .sort_values('cost',ascending=False)\
    .head()
```

```python
# Q2 Which room does make the most amount of income?
df.groupby('room_id').agg({'cost':'sum'})\
    .sort_values('cost',ascending=False)\
    .head()
```

```python
# Q3 How many time Jack Jones rent the room?
df.groupby('name').agg({'room_id':'count'})\
    .query('name=="Jack Jones"')
```

```python
# Q4 What are the rooms rented by Claire Taylor?
df.groupby(['name','room_id']).agg({'cost':'sum'})\
    .query('name=="Claire Taylor"')
```

```python
# Q5 what is the total income of ALL rooms in June? Between 1st June 2018(inclusive) and 30th June 2018(inclusive), (use return date).
# Q6 What are the total incomes of EACH room in June? Between 1st June 2018(inclusive) and 30th June 2018(inclusive), (use return date)
# Q7 Find members that have total rent time exactly 8 hours in June?
# Q8 Find members that have total rent time greater than or equal than 9 hours in June?
# Q9 Find the total rent duration of EACH room in June?
# Q10 Sort room (from max to min) by the total income in June
```

```python
from datetime import datetime, timedelta
df['hours']=(df_rent.stop-df_rent.start).apply(lambda x: x.seconds)//3600
init = datetime.strptime( "2018-06-01T00:00:00", "%Y-%m-%dT%H:%M:%S" )
final = datetime.strptime( "2018-06-30T23:59:59", "%Y-%m-%dT%H:%M:%S" )
df2=df[(init<=df.stop) & (df.stop<=final)]
```

```python
df2.sort_values('stop').head()
```

```python
df2.sort_values('stop').tail()
```

```python
# Q5 what is the total income of ALL rooms in June? Between 1st June 2018(inclusive) and 30th June 2018(inclusive), (use return date).
df2.sum(axis=0).cost
```

```python
# Q6 What are the total incomes of EACH room in June? Between 1st June 2018(inclusive) and 30th June 2018(inclusive), (use return date)
df2.groupby(['room_id']).agg({'cost':'sum'})
```
